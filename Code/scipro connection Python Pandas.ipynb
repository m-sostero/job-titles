# Connecting to Scipro Oracle database using Pandas and SQLalchemy

```python
from sqlalchemy.engine import create_engine
import pandas as pd

# Suppress scientific notation when printing DataFrame objects.
# We choose to display floats with 5 decimals. Modify as needed.
pd.options.display.float_format = '{:.5f}'.format
```

## Enter connection parameters; compose connection path

```python
DIALECT = 'oracle'
SQL_DRIVER = 'cx_oracle'
USERNAME = 'READ_BGLASS' #enter your username
PASSWORD = '' #enter your password
HOST = 's-jrcsvqdb121p.net1.cec.eu.int' #enter the oracle db host url
PORT = 1521 # enter the oracle port number
SERVICE = 'scipro' # enter the oracle db service name

ENGINE_PATH_WIN_AUTH = DIALECT + '+' + SQL_DRIVER + '://' + USERNAME + ':' + PASSWORD +'@' + HOST + ':' + str(PORT) + '/?service_name=' + SERVICE
```

## Create connection engine. 

Note that Oracle version 19 is known to have a maximum identifier length of 128, rather than the historical default of 30, set parameter accordingly.

```python
engine = create_engine(ENGINE_PATH_WIN_AUTH, max_identifier_length=128)
```

## Test connection to remote table on database

```python
test_df = pd.read_sql_query('SELECT * FROM "OWN_BGLASS"."SKILLS_UNIQUE"', engine)
test_df
```
